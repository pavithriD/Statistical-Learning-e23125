#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



1. Probability Space $(\Omega, \mathcal{F}, P)$

A suitable probability space for this discrete model is:

Sample Space ($\Omega$): The set of all possible outcomes.
$$\Omega = \{0, 1, 2, 3\}$$

Sigma-algebra ($\mathcal{F}$): The collection of all possible events (subsets of $\Omega$). For a finite set, we use the power set.
$$\mathcal{F} = \mathcal{P}(\Omega)$$

Probability Measure ($P$): A function $P: \mathcal{F} \to [0, 1]$ that assigns probabilities to events according to the given values:
$$P(\{0\}) = 0.50, \quad P(\{1\}) = 0.30, \quad P(\{2\}) = 0.15, \quad P(\{3\}) = 0.05$$

2. Definition of Random Variable $X$

We define $X$ as the identity map $X: \Omega \to \mathbb{R}$ such that $X(\omega) = \omega$ for all $\omega \in \Omega$. This maps the outcomes directly to the number of failures.

3. Probability Mass Function (PMF)

The PMF $p_X(x) = P(X = x)$ is:

$X = 0$  $p_X(x) = 0.50$

$X = 1$  $p_X(x) = 0.30$

$X = 2$  $p_X(x) = 0.15$

$X = 3$  $p_X(x) = 0.05$

4. Expectation $E[X]$

The expected value is the weighted average of all possible outcomes:

$$E[X] = \sum x \cdot P(X=x)$$

$$E[X] = (0 \times 0.50) + (1 \times 0.30) + (2 \times 0.15) + (3 \times 0.05)$$

$$E[X] = 0 + 0.30 + 0.30 + 0.15 = \mathbf{0.75}$$

5. Variance $Var(X)$

$$E[X^2] = (0^2 \times 0.50) + (1^2 \times 0.30) + (2^2 \times 0.15) + (3^2 \times 0.05)$$

$$E[X^2] = 0 + 0.30 + 0.60 + 0.45 = 1.35$$

$$Var(X) = E[X^2] - (E[X])^2$$

$$Var(X) = 1.35 - (0.75)^2 = 1.35 - 0.5625 = \mathbf{0.7875}$$

6. Probabilities $P(X \ge 1)$ and $P(X \ge 2)$

$$P(X \ge 1) = 1 - P(X=0) = 1 - 0.50 = \mathbf{0.50}$$

$$P(X \ge 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$$

7. Expectation and Variance Practical Meaning

Expectation ($E[X] = 0.75$): This represents the long-term average number of emergency transformer failures per day. If you monitored this district over many years, you would expect an average of 0.75 failures daily.

Variance ($Var(X) = 0.7875$): This measures the uncertainty or spread of the failure reports. Since the variance is relatively high compared to the mean (the standard deviation $\approx 0.887$ is larger than the mean), it suggests that while "0 failures" is most common, there is significant daily volatility in how many failures occur.


# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


1. Probability Space $(\Omega, \mathcal{F}, P)$

Sample Space ($\Omega$): The set of all possible proportions, $\Omega = [0, 1]$.

$\sigma$-field ($\mathcal{F}$): The Borel $\sigma$-algebra on $[0, 1]$, denoted as $\mathcal{B}([0, 1])$. This contains all intervals and their countable unions/intersections within the sample space.

Probability Measure ($P$): For any event $A \in \mathcal{F}$, the probability is defined by the integral of the density function over that set:
$P(A) = \int_A f(x) \, dx$.

2. Definition of $X$

The random variable $X$ is defined as the identity map $X: \Omega \to [0, 1]$, where $X(\omega) = \omega$. Since the mapping is continuous, it is measurable with respect to $\mathcal{F}$.

3. Probability Density Function

Given $\alpha = 2$ and $\beta = 18$, the PDF is:

$$f(x) = \frac{x^{2-1}(1-x)^{18-1}}{B(2, 18)} = \frac{x(1-x)^{17}}{B(2, 18)}, \quad 0 < x < 1$$

And $f(x) = 0$ otherwise.

4. Verification of the PDF

To be a valid PDF, $f(x)$ must satisfy two conditions:

Non-negativity: For $x \in (0, 1)$, $x > 0$ and $(1-x) > 0$. Since the Beta function $B(2, 18)$ is positive, $f(x) \geq 0$.

Normalization: The total area must equal 1.

$$\int_0^1 \frac{x(1-x)^{17}}{B(2, 18)} \, dx = \frac{1}{B(2, 18)} \int_0^1 x^{2-1}(1-x)^{18-1} \, dx$$

By the definition of the Beta function, $\int_0^1 x^{\alpha-1}(1-x)^{\beta-1} \, dx = B(\alpha, \beta)$.
Thus, $\frac{B(2, 18)}{B(2, 18)} = 1$.

5. Expectation $E[X]$

For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$:

$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$

6. Variance $Var(X)$

The formula for variance is $Var(X) = \frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$:

$$Var(X) = \frac{(2)(18)}{(20)^2(21)} = \frac{36}{400 \times 21} = \frac{36}{8400} = \frac{3}{700} \approx 0.004286$$

7. $P(X \leq 0.10)$

$$P(X \leq 0.10) = \frac{1}{B(2, 18)} \int_0^{0.10} x(1-x)^{17} \, dx$$

8. $P(0.05 \leq X \leq 0.15)$

$$P(0.05 \leq X \leq 0.15) = \frac{1}{B(2, 18)} \int_{0.05}^{0.15} x(1-x)^{17} \, dx$$

9. Suitability of the Beta Distribution

The Beta distribution is an excellent choice here because its support is strictly bounded within $[0, 1]$, matching the physical constraints of a proportion or percentage. Furthermore, it is highly flexible; by adjusting $\alpha$ and $\beta$, it can model various shapes (skewed, symmetric, or U-shaped), making it the standard conjugate prior for Bernoulli and Binomial processes in quality control.

10. Interpretation

Expectation ($E[X] = 0.10$): This represents the long-run average proportion of defective items. On average, 10% of the items produced daily are expected to be defective.

Variance ($Var(X) \approx 0.0043$): This measures the consistency of the production process. A low variance indicates that the daily defective rate stays very close to the 10% average, whereas a high variance would suggest that the process is unstable, with some days having very high or very low defect rates.

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


1. Support of $X$

The support of $X$ is the interval of all possible values for the installation time:

$$X \in [4, 10]$$

2. Probability Density Function (PDF)

$$f_X(x) = \begin{cases}
\frac{2(x-a)}{(b-a)(c-a)} & a \le x < c \
\frac{2(b-x)}{(b-a)(b-c)} & c \le x \le b \
0 & \text{otherwise}
\end{cases}$$

Plugging in $a=4, b=10, c=6$:

For $4 \le x < 6$: $\frac{2(x-4)}{(6)(2)} = \frac{x-4}{6}$

For $6 \le x \le 10$: $\frac{2(10-x)}{(6)(4)} = \frac{10-x}{12}$

3. Verification of Integration

To verify $\int_{4}^{10} f_X(x) dx = 1$:

$$\int_{4}^{6} \frac{x-4}{6} dx + \int_{6}^{10} \frac{10-x}{12} dx$$

$\left[ \frac{(x-4)^2}{12} \right]_4^6 = \frac{4}{12} - 0 = \frac{1}{3}$

$\left[ \frac{-(10-x)^2}{24} \right]_6^{10} = 0 - (-\frac{16}{24}) = \frac{2}{3}$

Total area = $\frac{1}{3} + \frac{2}{3} = 1$.

4. Expectation $E[X]$

The mean of a Triangular distribution is $\frac{a+b+c}{3}$:

$$E[X] = \frac{4+10+6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$

5. Variance $Var(X)$

The variance is given by $\frac{a^2+b^2+c^2-ab-ac-bc}{18}$:

$$Var(X) = \frac{4^2+10^2+6^2 - (40+24+60)}{18} = \frac{152 - 124}{18} = \frac{28}{18} \approx 1.56 \text{ hours}^2$$

6. Interpretation of Parameters

$a = 4$: The "best-case scenario." Even if everything goes perfectly, the installation will not take less than 4 hours.

$b = 10$: The "worst-case scenario." Barring extreme outliers, the installation should not exceed 10 hours.

$c = 6$: The "most likely" outcome. This is the peak of the distribution (the mode).

7. Why use a Triangular Distribution?

The Triangular distribution is ideal when data is scarce but expert knowledge is available. Unlike a Normal distribution, which requires a mean and standard deviation (often hard to estimate for unique projects), the Triangular model only requires intuitive bounds (min, max, mode), making it highly practical for construction and project management.

8. Probability $P(X \le 5)$

$$P(X \le 5) = \int_{4}^{5} \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_4^5 = \frac{1}{12} \approx 0.0833 \text{ (8.33\%)}$$

9. Probability $P(5 \le X \le 8)$

$$P(5 \le X \le 8) = P(5 \le X \le 6) + P(6 \le X \le 8)$$

$P(5 \le X \le 6) = \frac{1}{3} - \frac{1}{12} = \frac{3}{12} = 0.25$

$P(6 \le X \le 8) = \int_{6}^{8} \frac{10-x}{12} dx = \left[ \frac{-(10-x)^2}{24} \right]_6^8 = -\frac{4}{24} - (-\frac{16}{24}) = \frac{12}{24} = 0.5$

$$P(5 \le X \le 8) = 0.25 + 0.5 = 0.75 \text{ (75\%)}$$

10. Application in Project Planning

This model is a cornerstone of PERT (Program Evaluation and Review Technique).

Planning: It helps managers determine a realistic "buffer" for the hospital project.

Sensitivity Analysis: By shifting $b$ (e.g., if a part is delayed), managers can see how the mean completion time or the risk of missing a deadline (e.g., $P(X > 9)$) changes, allowing for better resource allocation.

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is given by

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


1. Shape Parameters $\alpha$ and $\beta$

Using the given formulas with $a=8$, $b=20$, and $c=11$:

$$\alpha = 1 + 4\left(\frac{11 - 8}{20 - 8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = 2$$

$$\beta = 1 + 4\left(\frac{20 - 11}{20 - 8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = 4$$

2. Support of $X$

The support is the interval of all possible values for the completion time:

$$S_X = [8, 20]$$

3. Probability Density Function $f_X(x)$

Substituting $\alpha=2$, $\beta=4$, $a=8$, and $b=20$:

$$f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2,4)(20-8)^{2+4-1}} = \frac{(x-8)(20-x)^3}{B(2,4) \cdot 12^5}$$

For $8 \le x \le 20$, and $0$ otherwise.

4. Probability Space $(\Omega, \mathcal{F}, P)$

Sample Space ($\Omega$): The set of all possible outcomes, $\Omega = [8, 20]$.

Sigma-algebra ($\mathcal{F}$): The Borel $\sigma$-algebra on $[8, 20]$, denoted $\mathcal{B}([8, 20])$, which contains all intervals and their completions.

Probability Measure ($P$): Defined by the integral of the PDF: $P(A) = \int_A f_X(x) \, dx$ for any $A \in \mathcal{F}$.

5. Definition of $X$ as a Random Variable

$X$ is a measurable function $X: \Omega \to \mathbb{R}$ such that $X(\omega) = \omega$. It maps each outcome in the sample space to a real number representing the number of days.

6. Expectation $E[X]$

The mean of a Beta distribution on $[0, 1]$ is $\mu_{beta} = \frac{\alpha}{\alpha + \beta}$. For a PERT distribution scaled to $[a, b]$, the mean is $a + (b-a)\frac{\alpha}{\alpha + \beta}$.

Substituting $\alpha = 1 + 4\frac{c-a}{b-a}$ and $\beta = 1 + 4\frac{b-c}{b-a}$:

$$\alpha + \beta = 2 + 4\left(\frac{c-a+b-c}{b-a}\right) = 2 + 4(1) = 6$$

$$E[X] = a + (b-a)\frac{1 + 4\frac{c-a}{b-a}}{6} = a + \frac{(b-a) + 4(c-a)}{6} = \frac{6a + b - a + 4c - 4a}{6} = \frac{a + 4c + b}{6}$$

For this project: $E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = 12$ days.

7. Variance $Var(X)$

$$Var(X) = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \cdot 8}{7} = \frac{32}{7} \approx 4.57 \text{ days}^2$$

8. Standard Deviation $\sigma$

$$\sigma = \sqrt{Var(X)} = \sqrt{\frac{32}{7}} \approx 2.14 \text{ days}$$

9. PERT vs. Triangular Distribution

A Triangular distribution with these parameters would have a mean of $\frac{8+11+20}{3} = 13$ days.The PERT distribution ($12$ days) is more appropriate here because it is "smoother" and places less emphasis on the extreme values ($a$ and $b$) compared to the Triangular model. In software integration, extreme delays are possible but usually less probable than the expert's "most likely" estimate suggests. PERT behaves more like a Normal distribution, providing a more realistic "bell-like" curve that centers more tightly around the most likely value $c$.

10. Practical Interpretation

$a$ (8 days): The "Best Case" scenario; assumes everything goes perfectly with no bugs or delays.

$b$ (20 days): The "Worst Case" scenario; accounts for significant integration hurdles or technical debt.

$c$ (11 days): The "Mode"; the duration seen most often in similar past tasks.

$E[X]$ (12 days): The "Risk-Adjusted" average; the value a manager should use for scheduling to account for the skew towards the longer duration.

$\sigma$ (2.14 days): The "Uncertainty"; a measure of risk. A larger $\sigma$ indicates a less predictable task. Managers often use $E[X] + 2\sigma$ to set "safe" deadlines.
