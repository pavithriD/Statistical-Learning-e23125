# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


1. Uniform Distribution with Respect to Area

In this model, the probability is determined by the ratio of the area of the region of interest to the total area of the disk.

Total Area ($S$):

$$S = \pi \cdot 10^2 = 100\pi$$

Area of Event $A$ ($S_A$):

The event $A$ covers the region where the radius is between $9$ and $10$.

$$S_A = \pi(10^2) - \pi(9^2) = 100\pi - 81\pi = 19\pi$$

Probability $P(A)$:

$$P(A) = \frac{S_A}{S} = \frac{19\pi}{100\pi} = 0.19$$


2. Uniform Distribution with Respect to Radius

In this model, the distance $R$ is chosen uniformly from the interval $[0, 10]$, meaning the probability depends only on the length of the radial interval.

Probability Density Function (PDF) of $R$:

$$f(r) = \frac{1}{10}, \quad 0 \le r \le 10$$

Probability $P(A)$:

Event $A$ occurs when $9 \le R \le 10$.

$$P(A) = \int_{9}^{10} f(r) \, dr = \int_{9}^{10} \frac{1}{10} \, dr = \left[ \frac{r}{10} \right]_{9}^{10} = \frac{10}{10} - \frac{9}{10} = 0.1$$

3. Comparison of the Two Models

The two answers differ because they represent different physical interpretations of "random":

Model 1 (Area): Assumes every $dx \, dy$ unit of area is equally likely. Because the "outer rings" of a disk have more surface area than the "inner rings," the point is more likely to be near the boundary.

Model 2 (Radius): Assumes every $dr$ unit of distance from the center is equally likely. This effectively "stretches" the probability density toward the center, making a point near the middle just as likely as a point near the edge, even though the middle has much less actual area.


# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

1. The "Random Endpoints" Method

In this model, we fix one endpoint of the chord and choose the second endpoint uniformly at random along the circumference.

Logic: Let the first endpoint be a vertex of the inscribed equilateral triangle. The circumference is divided into three equal arcs by the triangle's vertices.

Condition: For the chord to be longer than the side length $\sqrt{3}R$, the second endpoint must fall on the arc directly opposite the first vertex (between the other two vertices).

Calculation: This arc represents exactly $1/3$ of the total circumference.

Probability: $P = \frac{1}{3} \approx 0.333$

2. The "Random Radial Distance" Method

In this model, we choose a radius of the circle at random and then choose a point on that radius at random to be the midpoint of the chord.

Logic: A chord is uniquely determined by its midpoint. If we pick a random radius, the side of the inscribed triangle intersects that radius at exactly $R/2$ (half the distance from the center).

Condition: For the chord to be longer than the side length of the triangle, its midpoint must be closer to the center than the side of the triangle.

Calculation: The midpoint must lie on the segment of the radius between the center ($0$) and the distance $R/2$.

Probability: $P = \frac{R/2}{R} = \frac{1}{2} = 0.5$

3. The "Random Midpoint" Method

In this model, we choose a point anywhere inside the circle at random and treat it as the midpoint of the chord.

Logic: To define a chord, we pick any point $(x, y)$ within the circle of area $\pi R^2$.

Condition: A chord is longer than the side of the inscribed triangle if its midpoint falls within a smaller concentric circle that is tangent to the triangle's sides. The radius of this inner circle is $R/2$.

Calculation: The area of this inner circle is $\pi (R/2)^2 = \frac{1}{4} \pi R^2$.

Probability: $P = \frac{\text{Area of inner circle}}{\text{Area of outer circle}} = \frac{1/4 \pi R^2}{\pi R^2} = \frac{1}{4} = 0.25$

The "paradox" exists because the term "at random" is ambiguous. Depending on the physical process used to generate the chord (e.g., throwing a needle vs. picking points on a map), different distributions emerge. In modern probability, this is resolved by the Maximum Ignorance Principle, which often favors Method 2 ($1/2$) due to its rotational and translational invariance.

What is the probability that both children are girls, given that **at least one** of them is a girl?

The problem states that at least one of the children is a girl. We must filter our sample space to include only the outcomes that satisfy this condition:$\{B, G\}$$\{G, B\}$$\{G, G\}$

This gives us a reduced sample space of 3 possible outcomes.

From this reduced set, we want to find the probability that both children are girls. Only one outcome fits this description:$\{G, G\}$

Using the formula for conditional probability $P(A|B) = \frac{P(A \cap B)}{P(B)}$:$$P(\text{Both } G \mid \text{At least one } G) = \frac{1}{3}$$So, the probability is 1/3 (or approximately 33.3%).

# Q9) Information Available in a Discrete Random Variable

A factory inspects a product and records its quality using two different coding systems.

Each product falls into exactly one of the following four categories:

* $G_1$: good quality, produced on Machine 1
* $G_2$: good quality, produced on Machine 2
* $D_1$: defective, produced on Machine 1
* $D_2$: defective, produced on Machine 2

Assume the sample space is

$$
\Omega=\{G_1,G_2,D_1,D_2\},
$$

with probabilities

$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.

Answer the following.

4. Show that $\sigma(X)\subseteq \sigma(Y)$, and explain why $Y$ contains more information than $X$.

5. Find the marginal distribution of $X$.

6. Find the marginal distribution of $Y$.

   $$
   P(X=0,Y=1),\quad P(X=0,Y=2),\quad P(X=1,Y=3),\quad P(X=1,Y=4).
   $$

   $$
   P(X=1\mid Y=3),\qquad P(X=1\mid Y=4),\qquad P(Y=3\mid X=1),\qquad P(Y=4\mid X=1).
   $$

11. Give a Python based visualization of the model.

12. Find the marginal, joint, and conditional Shannon Entropy of $X$ and $Y$.

1. The Probability Space $(\Omega, \mathcal{F}, P)$

The sample space $\Omega$ represents the four possible outcomes. The $\sigma$-algebra $\mathcal{F}$ is the power set $\mathcal{P}(\Omega)$, containing all $2^4 = 16$ possible subsets of $\Omega$.

Sample Space: $\Omega = \{G_1, G_2, D_1, D_2\}$

Probability Measure:

$P(\{G_1\}) = 0.50$

$P(\{G_2\}) = 0.20$

$P(\{D_1\}) = 0.10$

$P(\{D_2\}) = 0.20$

2. $\sigma$-algebras Generated by $X$ and $Y$

A $\sigma$-algebra generated by a random variable represents the "information" we have if we only know the value of that variable.

$\sigma(X)$: Since $X$ maps $\{G_1, G_2\} \to 0$ and $\{D_1, D_2\} \to 1$, the atoms are $A_0 = \{G_1, G_2\}$ and $A_1 = \{D_1, D_2\}$.$$\sigma(X) = \{\emptyset, \Omega, \{G_1, G_2\}, \{D_1, D_2\}\}$$$\sigma(Y)$: Since $Y$ assigns a unique value to every single outcome in $\Omega$, it generates the full power set.$$\sigma(Y) = \mathcal{P}(\Omega) = \{\emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \dots, \Omega\}$$

3. Comparison of Information

Proof of Inclusion: Every set in $\sigma(X)$ is also in $\sigma(Y)$. For example, $\{G_1, G_2\} \in \sigma(Y)$ because it is the union of $\{G_1\}$ and $\{G_2\}$. Thus, $\sigma(X) \subseteq \sigma(Y)$.

Interpretation: $Y$ contains more information because it distinguishes between machines ($1$ vs $2$). Knowing $Y$ tells you both the quality and the machine. Knowing $X$ only tells you the quality; you lose the information regarding which machine produced the item.

4. Marginal Distributions

$P(\{G_1, G_2\}) = 0.7$

$P(\{D_1, D_2\}) = 0.3$

$P(\{G_1\}) = 0.5$

$P(\{G_2\}) = 0.2$

$P(\{D_1\}) = 0.1$

$P(\{D_2\}) = 0.2$

5. Joint and Conditional Probabilities

Joint Probabilities:

Since $Y$ is a refinement of $X$, the joint probabilities are simply the probabilities of the individual outcomes:

$P(X=0, Y=1) = 0.50$

$P(X=0, Y=2) = 0.20$

$P(X=1, Y=3) = 0.10$

$P(X=1, Y=4) = 0.20$

Conditional Probabilities:

$P(X=1 \mid Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.1}{0.1} = \mathbf{1.0}$$P(X=1 \mid Y=4) = \frac{P(X=1, Y=4)}{P(Y=4)} = \frac{0.2}{0.2} = \mathbf{1.0}$$P(Y=3 \mid X=1) = \frac{P(X=1, Y=3)}{P(X=1)} = \frac{0.1}{0.3} = \mathbf{1/3}$$P(Y=4 \mid X=1) = \frac{P(X=1, Y=4)}{P(X=1)} = \frac{0.2}{0.3} = \mathbf{2/3}$

Conditioning on $Y$ leaves no uncertainty about $X$ (if you know it's Machine 1 and Defective, you definitely know it's Defective). However, conditioning on $X$ still leaves uncertainty about $Y$ (if you know it's Defective, you still don't know for sure which machine made it).

6. Information Discrepancy

In $\sigma(X)$ but not $\sigma(Y)$: Strictly speaking, nothing. Because $\sigma(X) \subseteq \sigma(Y)$, any information $X$ provides is already present in $Y$.

In $\sigma(Y)$ but not $\sigma(X)$: The Machine ID. $Y$ allows us to distinguish between $G_1$ and $G_2$. In $\sigma(X)$, these are lumped together into a single event "Good Quality."

7. Shannon Entropy Calculation

Using the formula $H(Z) = -\sum p_i \log_2(p_i)$:

Marginal $H(X)$:$H(X) = -(0.7 \log_2 0.7 + 0.3 \log_2 0.3) \approx \mathbf{0.881 \text{ bits}}$

Marginal $H(Y)$:$H(Y) = -(0.5 \log_2 0.5 + 2(0.2 \log_2 0.2) + 0.1 \log_2 0.1) \approx \mathbf{1.761 \text{ bits}}$

Joint $H(X, Y)$:

Since $Y$ determines $X$, $H(X, Y) = H(Y) \approx \mathbf{1.761 \text{ bits}}$.

Conditional $H(Y|X)$:$H(Y|X) = H(X, Y) - H(X) = 1.761 - 0.881 = \mathbf{0.880 \text{ bits}}$. (This represents the remaining uncertainty about the machine after knowing the quality).

A chemical plant monitors each production run using two different measurement systems.

  $$
  X(t,p)=t,

   $$

h(Y)=-\int_{\mathbb R^2} f_Y(t,p)\ln f_Y(t,p)\,dt\,dp.

1. The Probability Space $(\Omega, \mathcal{F}, P)$

Sample Space ($\Omega$): $\mathbb{R}^2$. This represents all possible pairs of $(t, p)$.

$\sigma$-algebra ($\mathcal{F}$): $\mathcal{B}(\mathbb{R}^2)$, the Borel $\sigma$-algebra on $\mathbb{R}^2$.

Probability Measure ($P$): For any set $A \in \mathcal{F}$:

$$P(A) = \iint_A \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right) dt dp$$

2. The $\sigma$-algebra generated by $X$, $\sigma(X)$

The variable $X(t, p) = t$ only "sees" the first coordinate.

$$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathcal{B}(\mathbb{R}) \}$$

This consists of "vertical strips" in $\mathbb{R}^2$. Once you know which strip you are in, you know $t$, but you have no information about $p$.

3. The $\sigma$-algebra generated by $Y$, $\sigma(Y)$

Since $Y(t, p) = (t, p)$ is the identity map on $\mathbb{R}^2$:

$$\sigma(Y) = \mathcal{B}(\mathbb{R}^2)$$

It contains every Borel-measurable subset of the plane.

4. Comparison of $\sigma(X)$ and $\sigma(Y)$

Inclusion: Every set in $\sigma(X)$ (a vertical strip) is by definition a Borel set in $\mathbb{R}^2$, so $\sigma(X) \subseteq \sigma(Y)$.

Information: $Y$ contains more information because $\sigma(Y)$ is "finer." While $\sigma(X)$ can only distinguish between events based on $t$, $\sigma(Y)$ can distinguish between events based on both $t$ and $p$. $Y$ tells you the exact point; $X$ only tells you the vertical line that point lies on.

5. Marginal Density of $X$

Since $X$ is just $T$, and $T \sim N(0, 1)$:

$$f_X(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)$$

6. Marginal Density of $Y$

$Y$ is the joint vector $(T, P)$. Its density is the joint density provided:

$$f_Y(t, p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$$

7. Probabilities

$P(X \le 0)$: Since $X \sim N(0, 1)$, $P(X \le 0) = 0.5$.

$P(X > 1)$: $1 - \Phi(1) \approx 1 - 0.8413 = 0.1587$.

$P(Y \in (-\infty, 0] \times \mathbb{R})$: This is the same as $P(T \le 0)$, which is $0.5$.

$P(Y \in [-1, 1] \times [-2, 2])$: Due to independence:

$$P(-1 \le T \le 1) \times P(-2 \le P \le 2)$$

For $T$: $\Phi(1) - \Phi(-1) \approx 0.6827$.

For $P \sim N(0, 4)$, the standard deviation $\sigma = 2$. So $P(-2 \le P \le 2)$ is $P(-1\sigma \le P \le 1\sigma) \approx 0.6827$.

Result: $0.6827 \times 0.6827 \approx \mathbf{0.4661}$.

8. Conditional Distribution of $X$ given $Y=(t, p)$

If we know $Y=(t, p)$, we know the exact values of both $T$ and $P$. Therefore, $X$ is no longer random; it is exactly $t$.

$$X | Y=(t, p) \sim \delta_t \text{ (Dirac delta distribution at } t)$$

9. Conditional Distribution of $Y$ given $X=t$

Since $T$ and $P$ are independent, knowing $T=t$ tells us nothing about $P$.

$$Y | X=t \sim (t, P) \text{ where } P \sim N(0, 4)$$

The conditional density is $f_{Y|X}(t, p | t) = \frac{1}{\sqrt{8\pi}} \exp\left(-\frac{p^2}{8}\right)$.

10. Information Summary

Conditioning on $X$: We know the temperature deviation. We know we are on a specific vertical line in the state space, but we are uncertain about the pressure (the vertical position on that line).

Conditioning on $Y$: All uncertainty is removed. We know both the temperature and the pressure deviations.

11. Measurability

Since $X(t, p) = t$ and $Y(t, p) = (t, p)$, we have $X = \pi_1 \circ Y$. Because $\pi_1$ (the projection map) is continuous, it is Borel-measurable. A measurable function of a random variable is itself a random variable measurable with respect to the $\sigma$-algebra of the source ($Y$).

13 & 14. Differential Entropy

The entropy of a normal distribution $N(\mu, \sigma^2)$ is $h = \frac{1}{2} \ln(2\pi e \sigma^2)$.

$h(X)$: $\sigma^2 = 1 \implies h(X) = \frac{1}{2} \ln(2\pi e) \approx 1.4189$ nats.

$h(Y)$: Since components are independent, $h(Y) = h(T) + h(P)$.$h(P)$ has $\sigma^2 = 4 \implies h(P) = \frac{1}{2} \ln(2\pi e \cdot 4) = h(X) + \ln(2) \approx 2.1121$.$h(Y) = 1.4189 + 2.1121 = \mathbf{3.5310}$ nats.

15. Conditional Differential Entropy $h(Y|X)$

Because $T$ and $P$ are independent:

$$h(Y|X) = h(T, P | T) = h(P | T) = h(P) = \mathbf{2.1121} \text{ nats.}$$

Physical Interpretation: This represents the remaining uncertainty in the system state (the pressure) once the temperature is already known.

16. Discussion on $h(X|Y)$

In this model, $X$ is a deterministic function of $Y$.

Discrete Case: If $X = g(Y)$, then $H(X|Y) = 0$.

Continuous Case: $h(X|Y)$ formally tends to $-\infty$. This is because knowing $Y$ identifies $X$ with infinite precision (a Dirac delta). Differential entropy can be negative and behaves differently than discrete entropy; the $-\infty$ signifies that there is "zero volume" of uncertainty remaining.
