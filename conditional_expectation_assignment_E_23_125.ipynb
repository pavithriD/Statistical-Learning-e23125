X(\omega_1)=0,\qquad

Y(\omega_2)=0,\qquad

1. The Probability Space $(\Omega, \mathcal{F}, P)$

A probability space consists of a sample space $\Omega$, a $\sigma$-algebra $\mathcal{F}$, and a probability measure $P$.

Sample Space ($\Omega$):
$$\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$$

$\sigma$-algebra ($\mathcal{F}$): Since $\mathcal{F} = \mathcal{P}(\Omega)$ (the power set), it contains all $2^4 = 16$ possible subsets of $\Omega$:
$$\mathcal{F} = \big\{ \emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1, \omega_2\}, \{\omega_1, \omega_3\}, \{\omega_1, \omega_4\}, \{\omega_2, \omega_3\}, \{\omega_2, \omega_4\}, \{\omega_3, \omega_4\}, \{\omega_1, \omega_2, \omega_3\}, \{\omega_1, \omega_2, \omega_4\}, \{\omega_1, \omega_3, \omega_4\}, \{\omega_2, \omega_3, \omega_4\}, \Omega \big\}$$

Probability Measure ($P$): Defined on the singleton events as given:
$$P(\{\omega_1\}) = 0.50, \quad P(\{\omega_2\}) = 0.10, \quad P(\{\omega_3\}) = 0.20, \quad P(\{\omega_4\}) = 0.20$$

2. The $\sigma$-algebra Generated by $Y$, $\sigma(Y)$

The random variable $Y$ groups the outcomes based on the values it maps to:

For $Y = 0$: $Y^{-1}(\{0\}) = \{\omega_1, \omega_2\}$

For $Y = 1$: $Y^{-1}(\{1\}) = \{\omega_3, \omega_4\}$

The $\sigma$-algebra generated by $Y$ is the collection of pre-images of all Borel sets under $Y$, which consists of the empty set, the full space, and these two preimage blocks:
$$\sigma(Y) = \big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \big\}$$

3. Measurability of $E[X \mid Y]$

By definition, the conditional expectation $E[X \mid Y]$ is a random variable that is $\sigma(Y)$-measurable.
For a random variable to be $\sigma(Y)$-measurable, the pre-image of any Borel set must lie in $\sigma(Y)$. Practically, this means $E[X \mid Y](\omega)$ cannot distinguish between outcomes that lie within the same partition block of $\sigma(Y)$.
Because the atoms (the smallest non-empty building blocks) of $\sigma(Y)$ are exactly $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, $E[X \mid Y]$ cannot change values within these sets. It must remain constant across all $\omega \in \{\omega_1, \omega_2\}$ and constant across all $\omega \in \{\omega_3, \omega_4\}$.

4. Explicit Computation of $E[X \mid Y]$

Since $E[X \mid Y]$ is constant on the atoms of $\sigma(Y)$, we evaluate it standardly as a piecewise function over the events $\{Y=0\}$ and $\{Y=1\}$:

On the set $\{Y=0\} = \{\omega_1, \omega_2\}$:
$$E[X \mid Y=0] = \frac{E[X \cdot \mathbb{1}_{\{Y=0\}}]}{P(Y=0)} = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(\{\omega_1, \omega_2\})}$$
$$E[X \mid Y=0] = \frac{(0)(0.50) + (1)(0.10)}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$

On the set $\{Y=1\} = \{\omega_3, \omega_4\}$:
$$E[X \mid Y=1] = \frac{E[X \cdot \mathbb{1}_{\{Y=1\}}]}{P(Y=1)} = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(\{\omega_3, \omega_4\})}$$
$$E[X \mid Y=1] = \frac{(0)(0.20) + (1)(0.20)}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$

Combining these yields the requested indicator function expression:
$$E[X \mid Y] = \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}$$

5. Verification of the Defining Property

We must check that $\int_A E[X \mid Y] dP = \int_A X dP$ for all four sets $A \in \sigma(Y)$:

For $A = \emptyset$: Both sides vacuously integrate to $0$.

For $A = \Omega$:
$\int_\Omega X dP = E[X] = (0)(0.50) + (1)(0.10) + (0)(0.20) + (1)(0.20) = 0.30$
$\int_\Omega E[X \mid Y] dP = \left(\frac{1}{6}\right)P(Y=0) + \left(\frac{1}{2}\right)P(Y=1) = \left(\frac{1}{6}\right)(0.60) + \left(\frac{1}{2}\right)(0.40) = 0.10 + 0.20 = 0.30$

For $A = \{\omega_1, \omega_2\}$ (i.e., $\{Y=0\}$):
$\int_A X dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0 + 0.10 = 0.10$
$\int_A E[X \mid Y] dP = \frac{1}{6} P(\{\omega_1, \omega_2\}) = \frac{1}{6}(0.60) = 0.10$

For $A = \{\omega_3, \omega_4\}$ (i.e., $\{Y=1\}$):
$\int_A X dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0 + 0.20 = 0.20$
$\int_A E[X \mid Y] dP = \frac{1}{2} P(\{\omega_3, \omega_4\}) = \frac{1}{2}(0.40) = 0.20$

The property holds perfectly across $\sigma(Y)$.

6. Conditional PMF $p_{X \mid Y}(x \mid y)$

Using the joint and marginal probabilities calculated from the sample space:

$P(X=0, Y=0) = P(\{\omega_1\}) = 0.50$$P(X=1, Y=0) = P(\{\omega_2\}) = 0.10$$P(X=0, Y=1) = P(\{\omega_3\}) = 0.20$$P(X=1, Y=1) = P(\{\omega_4\}) = 0.20$

We compute $p_{X \mid Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$:

$y = 0$  

$p_{X \mid Y}(0 \mid y) = \frac{0.50}{0.60} = \mathbf{\frac{5}{6}}$

$p_{X \mid Y}(1 \mid y) = \frac{0.10}{0.60} = \mathbf{\frac{1}{6}}$


$y = 1$  

$p_{X \mid Y}(0 \mid y) = \frac{0.20}{0.40} = \mathbf{\frac{1}{2}}$

$p_{X \mid Y}(1 \mid y) = \frac{0.20}{0.40} = \mathbf{\frac{1}{2}}$

7. Identity Verification

Let's test the traditional discrete expectation formula $\sum x \cdot p_{X \mid Y}(x \mid y)$:

For $y = 0$:$$\sum_{x \in \{0,1\}} x \cdot p_{X \mid Y}(x \mid 0) = 0 \cdot \left(\frac{5}{6}\right) + 1 \cdot \left(\frac{1}{6}\right) = \frac{1}{6}$$
This matches $E[X \mid Y=0] = \frac{1}{6}$.

For $y = 1$:$$\sum_{x \in \{0,1\}} x \cdot p_{X \mid Y}(x \mid 1) = 0 \cdot \left(\frac{1}{2}\right) + 1 \cdot \left(\frac{1}{2}\right) = \frac{1}{2}$$This matches $E[X \mid Y=1] = \frac{1}{2}$.

8. Mean Square Error (MSE) Using $Y$

The MSE of the conditional expectation estimator $E[X \mid Y]$ is equivalent to the Expected Conditional Variance:
$$\text{MSE} = E\left[ (X - E[X \mid Y])^2 \right] = E[\text{Var}(X \mid Y)]$$

First, compute $\text{Var}(X \mid Y = y)$ for each state:
$\text{Var}(X \mid Y=0) = \left(\frac{1}{6}\right)\left(1 - \frac{1}{6}\right) = \frac{5}{36}$
$\text{Var}(X \mid Y=1) = \left(\frac{1}{2}\right)\left(1 - \frac{1}{2}\right) = \frac{1}{4} = \frac{9}{36}$

Now, compute the expected value over $Y$:$$\text{MSE} = \text{Var}(X \mid Y=0)P(Y=0) + \text{Var}(X \mid Y=1)P(Y=1)$$$$\text{MSE} = \left(\frac{5}{36}\right)(0.60) + \left(\frac{9}{36}\right)(0.40) = \frac{3}{36} + \frac{3.6}{36} = \frac{6.6}{36} = \frac{11}{60} \approx \mathbf{0.1833}$$

9. Omitting $Y$ vs. The Law of Total Variance

If the bank ignores $Y$ and uses the global average $E[X] = 0.30$, the new error is simply the total variance $\text{Var}(X)$. Because $X$ is a Bernoulli random variable with $p = 0.30$:$$\text{MSE}_{\text{constant}} = \text{Var}(X) = 0.30 \times (1 - 0.30) = \mathbf{0.2100}$$

Relationship to the Law of Total Variance:

The Law of Total Variance states:$$\text{Var}(X) = E[\text{Var}(X \mid Y)] + \text{Var}(E[X \mid Y])$$

Let's compute the variance of the estimator, $\text{Var}(E[X \mid Y])$:$E[E[X \mid Y]] = E[X] = 0.30$$\text{Var}(E[X \mid Y]) = P(Y=0)\left(\frac{1}{6} - 0.3\right)^2 + P(Y=1)\left(\frac{1}{2} - 0.3\right)^2$$\text{Var}(E[X \mid Y]) = 0.60\left(-\frac{2}{15}\right)^2 + 0.40(0.2)^2 = 0.60\left(\frac{4}{225}\right) + 0.40(0.04) = \frac{2.4}{225} + 0.016 = 0.01067 + 0.016 = 0.02667 = \frac{4}{150} = \frac{2}{75}$

Plugging these back into the law:$$\text{Var}(X) = \frac{11}{60} + \frac{2}{75} = 0.18333 + 0.02667 = 0.2100$$Explanation: Ignoring the risk class $Y$ increases the prediction error by exactly $\text{Var}(E[X \mid Y]) = 0.0267$. By observing $Y$, the bank successfully captures this variance component, effectively lowering its mean square uncertainty.

10. Information Theoretic Analysis (Entropy & Mutual Information)

a) Entropy $H(X)$Since $P(X=1) = 0.30$ and $P(X=0) = 0.70$:$$H(X) = -0.70 \log_2(0.70) - 0.30 \log_2(0.30) \approx 0.3653 + 0.5211 = \mathbf{0.8813 \text{ bits}}$$

b) Conditional Entropy $H(X \mid Y)$

$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$$H(X \mid Y=0) = -\frac{5}{6}\log_2\left(\frac{5}{6}\right) - \frac{1}{6}\log_2\left(\frac{1}{6}\right) \approx 0.2189 + 0.4310 = 0.6500 \text{ bits}$$H(X \mid Y=1) = -\frac{1}{2}\log_2\left(\frac{1}{2}\right) - \frac{1}{2}\log_2\left(\frac{1}{2}\right) = 1.0000 \text{ bits}$
$$H(X \mid Y) = 0.60(0.6500) + 0.40(1.0000) = 0.3900 + 0.4000 = \mathbf{0.7900 \text{ bits}}$$

c) Mutual Information $I(X; Y)$

$$I(X; Y) = H(X) - H(X \mid Y) = 0.8813 - 0.7900 = \mathbf{0.0913 \text{ bits}}$$

d) Percentage of Uncertainty Removed

$$\text{Percentage Removed} = \frac{I(X; Y)}{H(X)} \times 100\% = \frac{0.0913}{0.8813} \times 100\% \approx \mathbf{10.36\%}$$

Knowing the customer risk profile removes roughly $10.36\%$ of the bank's system uncertainty regarding fraud.






# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

To derive the conditional distribution $f_{X|Y}(x|y)$, we use the definition of conditional density:$$f_{X|Y}(x|y) = \frac{f_{X,Y}(x,y)}{f_Y(y)}$$The derivation relies on partitioning the joint quadratic form in the exponent and evaluating the determinant of the joint covariance matrix $\Sigma$. This is elegantly achieved using the Schur complement of $\Sigma_{YY}$.

Step 1: Matrix Inversion via Block Diagonalization

Let $S_{XX} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$ be the Schur complement of $\Sigma_{YY}$ in $\Sigma$. We can decompose the joint covariance matrix $\Sigma$ into a product of block matrices:$$\Sigma = \begin{bmatrix} I_n & \Sigma_{XY}\Sigma_{YY}^{-1} \\ 0 & I_m \end{bmatrix} \begin{bmatrix} S_{XX} & 0 \\ 0 & \Sigma_{YY} \end{bmatrix} \begin{bmatrix} I_n & 0 \\ \Sigma_{YY}^{-1}\Sigma_{YX} & I_m \end{bmatrix}$$

Taking the inverse of both sides yields:$$\Sigma^{-1} = \begin{bmatrix} I_n & 0 \\ -\Sigma_{YY}^{-1}\Sigma_{YX} & I_m \end{bmatrix} \begin{bmatrix} S_{XX}^{-1} & 0 \\ 0 & \Sigma_{YY}^{-1} \end{bmatrix} \begin{bmatrix} I_n & -\Sigma_{XY}\Sigma_{YY}^{-1} \\ 0 & I_m \end{bmatrix}$$

Multiplying these matrices gives the explicit block form of $\Sigma^{-1}$:$$\Sigma^{-1} = \begin{bmatrix} S_{XX}^{-1} & -S_{XX}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \\ -\Sigma_{YY}^{-1}\Sigma_{YX}S_{XX}^{-1} & \Sigma_{YY}^{-1} + \Sigma_{YY}^{-1}\Sigma_{YX}S_{XX}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \end{bmatrix}$$

Step 2: Decomposing the Joint Quadratic Form

The quadratic form in the exponent of the joint density is:$$Q = (z-\mu)^T \Sigma^{-1} (z-\mu)$$

Letting $\tilde{x} = x - \mu_X$ and $\tilde{y} = y - \mu_Y$, we can rewrite $Q$ by substituting our block representation of $\Sigma^{-1}$:

$$Q = \begin{bmatrix} \tilde{x}^T & \tilde{y}^T \end{bmatrix} \begin{bmatrix} S_{XX}^{-1} & -S_{XX}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \\ -\Sigma_{YY}^{-1}\Sigma_{YX}S_{XX}^{-1} & \Sigma_{YY}^{-1} + \Sigma_{YY}^{-1}\Sigma_{YX}S_{XX}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \tilde{x} \\ \tilde{y} \end{bmatrix}$$

Expanding this product and grouping terms associated with $S_{XX}^{-1}$ yields:$$Q = \tilde{x}^T S_{XX}^{-1} \tilde{x} - \tilde{x}^T S_{XX}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} - \tilde{y}^T\Sigma_{YY}^{-1}\Sigma_{YX}S_{XX}^{-1}\tilde{x} + \tilde{y}^T\Sigma_{YY}^{-1}\Sigma_{YX}S_{XX}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} + \tilde{y}^T\Sigma_{YY}^{-1}\tilde{y}$$

Notice that the first four terms perfectly factor into a single quadratic form centered around the conditional mean:$$Q = \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)^T S_{XX}^{-1} \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right) + \tilde{y}^T\Sigma_{YY}^{-1}\tilde{y}$$

Substituting back $\tilde{x}$ and $\tilde{y}$, we define:$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$\Sigma_{X|Y} = S_{XX} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$Thus, the joint quadratic form simplifies beautifully to:$$Q = (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*) + (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y)$$

Step 3: Determinant Decomposition

Using the properties of the determinant on our initial block factorization of $\Sigma$, the determinant of the triangular matrices is $1$. Therefore:$$|\Sigma| = \left| \begin{matrix} S_{XX} & 0 \\ 0 & \Sigma_{YY} \end{matrix} \right| = |S_{XX}| \cdot |\Sigma_{YY}| = |\Sigma_{X|Y}| \cdot |\Sigma_{YY}|$$

Step 4: Constructing the Conditional Density

We now substitute the expanded joint exponent and the factored determinant into the definition of the joint density $f_{X,Y}(x,y)$:

$$f_{X,Y}(x,y) = \frac{1}{(2\pi)^{\frac{n+m}{2}} \sqrt{|\Sigma_{X|Y}| \cdot |\Sigma_{YY}|}} \exp\left( -\frac{1}{2} (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*) \right) \exp\left( -\frac{1}{2} (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y) \right)$$

The marginal distribution of $Y$ is a standard multivariate normal vector of dimension $m$:$$f_Y(y) = \frac{1}{(2\pi)^{\frac{m}{2}} \sqrt{|\Sigma_{YY}|}} \exp\left( -\frac{1}{2} (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y) \right)$$

Dividing $f_{X,Y}(x,y)$ by $f_Y(y)$ scenario causes the terms depending strictly on $y$ to completely cancel out:
$$f_{X|Y}(x|y) = \frac{\frac{1}{(2\pi)^{\frac{n}{2}}(2\pi)^{\frac{m}{2}} \sqrt{|\Sigma_{X|Y}|}\sqrt{|\Sigma_{YY}|}} \exp\left(-\frac{1}{2} (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*)\right) \exp\left(-\frac{1}{2} (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y)\right)}{\frac{1}{(2\pi)^{\frac{m}{2}} \sqrt{|\Sigma_{YY}|}} \exp\left(-\frac{1}{2} (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y)\right)}$$

$$f_{X|Y}(x|y) = \frac{1}{(2\pi)^{\frac{n}{2}} \sqrt{|\Sigma_{X|Y}|}} \exp\left( -\frac{1}{2} (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*) \right)$$

Where:$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$

This completes the proof that the conditional distribution is an $n$-dimensional multivariate normal distribution, denoted as $X|Y=y \sim \mathcal{N}(\mu^*, \Sigma_{X|Y})$.

1. Joint Distribution

Identification of $(X, Y)$

Since $X$ and $Q$ are independent normal random variables, any linear combination of them is also normally distributed. Because $Y = X + Q$, the joint vector $(X, Y)$ follows a Bivariate Normal Distribution:$$(X, Y)^T \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$$

Show that $Y \sim \mathcal{N}(0, 5)$

Mean: By the linearity of expectation:$$E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$$

Variance: Because $X$ and $Q$ are independent, their variances add directly:$$Var(Y) = Var(X + Q) = Var(X) + Var(Q) = 1 + 4 = 5$$

Thus, the marginal distribution of the observation is $Y \sim \mathcal{N}(0, 5)$.

Find the Covariance $Cov(X, Y)$

Using the bilinear properties of covariance and the fact that $Cov(X, Q) = 0$ due to independence:$$Cov(X, Y) = Cov(X, X + Q) = Cov(X, X) + Cov(X, Q) = Var(X) + 0 = 1$$

2. Information Geometry & Measurability

Why $X$ is not $\sigma(Y)$-measurable

For a random variable $X$ to be $\sigma(Y)$-measurable, its value must be uniquely determined by the observation of $Y$. In other words, there must exist a deterministic function $g$ such that $X = g(Y)$.Here, $X$ is not $\sigma(Y)$-measurable because for any given value $Y = y$, there are infinitely many combinations of $X$ and $Q$ that can yield that exact sum.

Practical Interpretation

If the power meter displays $Y = 5$, the engineer faces an identifiability problem. This high reading could mean:

The machine is running dangerously hot ($X = 4$) while cutting regular material ($Q = 1$).

The machine is at a normal temperature ($X = 0$) but is cutting an exceptionally hard material ($Q = 5$).

Because the measurement aggregates two distinct physical phenomena into a single scalar value, the individual contributions are confounded.

3. Conditional Expectation (The Best Estimate)

Using the formula for conditional expectations of joint normal distributions:$$E[X \mid Y = y] = \mu_X + \frac{Cov(X,Y)}{Var(Y)}(y - \mu_Y)$$

Plugging in our calculated values ($\mu_X = 0$, $\mu_Y = 0$, $Cov(X,Y) = 1$, $Var(Y) = 5$):$$E[X \mid Y] = 0 + \frac{1}{5}(Y - 0) = \frac{1}{5}Y$$

Signal-to-Total-Variance Ratio

Expressing this in terms of the system's variances reveals how the estimator weighs the data:$$E[X \mid Y] = \left( \frac{Var(X)}{Var(X) + Var(Q)} \right) Y$$The coefficient $\frac{1}{5}$ is precisely the signal-to-total-variance ratio. It acts as a trust filter: if the noise variance $Var(Q)$ is high relative to the signal variance $Var(X)$, the system heavily discounts the measurement $Y$.

4. Verification of the Residual

Let $\hat{X} = E[X \mid Y] = \frac{1}{5}Y$. The estimation error is $Z = X - \hat{X} = X - \frac{1}{5}Y$.

$$E[Z] = E\left[X - \frac{1}{5}Y\right] = E[X] - \frac{1}{5}E[Y] = 0 - 0 = 0$$

$$Cov(Z, Y) = Cov\left(X - \frac{1}{5}Y, Y\right) = Cov(X, Y) - \frac{1}{5}Var(Y)$$

$$Cov(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$

Because the error is uncorrelated with our observation, it proves that $\hat{X}$ has extracted all possible linear information about $X$ contained within $Y$.

5. Information Theory: Differential Entropy

The differential entropy of a 1D Gaussian variable $W \sim \mathcal{N}(0, \sigma^2)$ is given by $h(W) = \frac{1}{2}\ln(2\pi e \sigma^2)$.

Differential Entropy $h(X)$

For $X \sim \mathcal{N}(0, 1)$:$$h(X) = \frac{1}{2}\ln(2\pi e (1)) = \frac{1}{2}\ln(2\pi e)$$

Conditional Differential Entropy $h(X \mid Y)$

First, we find the conditional variance $\sigma^2_{X|Y}$:$$\sigma^2_{X|Y} = Var(X) - \frac{Cov(X,Y)^2}{Var(Y)} = 1 - \frac{1^2}{5} = \frac{4}{5} = 0.8$$Because the conditional distribution $(X \mid Y = y)$ is normal and its variance is independent of the specific value $y$, the conditional entropy is:$$h(X \mid Y) = \frac{1}{2}\ln(2\pi e \sigma^2_{X|Y}) = \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right)$$

6. Mutual Information

Using the properties of logarithms:$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right)$$$$I(X; Y) = \frac{1}{2}\ln\left(\frac{1}{4/5}\right) = \frac{1}{2}\ln(1.25) \approx 0.1116 \text{ nats}$$

The entropy reduction represents the proportion of uncertainty removed:$$\text{Percentage Removed} = \frac{I(X;Y)}{h(X)} \cdot 100\%$$However, a more physically meaningful way to frame "uncertainty reduction" in variance terms is looking at how much variance was eliminated:$$\text{Variance Reduction} = \frac{Var(X) - Var(X|Y)}{Var(X)} = \frac{1 - 0.8}{1} = 20\%$$

If mechanical noise increases drastically to $Var(Q) = 100$:

$Var(Y)$ becomes $1 + 100 = 101$.$\sigma^2_{X|Y} = 1 - \frac{1^2}{101} = \frac{100}{101} \approx 0.9901$.$I(X; Y) = \frac{1}{2}\ln\left(\frac{1}{100/101}\right) = \frac{1}{2}\ln(1.01) \approx 0.0049 \text{ nats}$.

Conclusion: The mutual information drops drastically close to $0$. The power meter becomes virtually useless because the signal is completely drowned out by the mechanical load fluctuations.

7. Diagnostic Interpretation

Why $Y$ is a Biased Estimator of $X$

The manager's intuition fails because it ignores the presence of the noise variable $Q$.If we use $\hat{X}_{\text{manager}} = Y$, the expected value given a specific measurement $Y=y$ is simply $y$. However, the true conditional expectation is:$$E[X \mid Y = y] = 0.2y$$Using $Y$ directly as the estimate consistently overestimates the magnitude of the temperature deviation. It treats $100\%$ of the power spike as a thermal issue, forgetting that $80\%$ of the system's baseline variance stems from mechanical cutting loads ($4$ out of $5$ units of variance).

The "Shrinkage" Phenomenon

The conditional expectation formula $\frac{1}{5}Y$ explicitly shrinks the raw measurement toward the prior mean (zero).

This shrinkage is an elegant mathematical safeguard against noise. Because the power meter is noisy, any extreme reading (e.g., $Y=5$) is statistically more likely to be caused by a random spike in mechanical torque ($Q$) rather than a massive surge in internal temperature ($X$). By pulling the estimate back toward zero, Bayes' theorem optimizes the guess, balancing what the raw data claims against what the historical baselines dictate.
