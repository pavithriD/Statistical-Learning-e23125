1. Derivation of the Bias in $\hat{\Sigma}_{\text{MLE}}$

To show that the raw Maximum Likelihood Estimator $\hat{\Sigma}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} (X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T$ is biased, we must evaluate its mathematical expectation.

First, let's expand the term inside the summation. For convenience, let $\bar{X} = \hat{\mu}_n$. We can rewrite the deviation from the sample mean by injecting the true population mean $\mu$:

$$X_i - \bar{X} = (X_i - \mu) - (\bar{X} - \mu)$$

$$\sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T = \sum_{i=1}^{n} \left[ ((X_i - \mu) - (\bar{X} - \mu))((X_i - \mu) - (\bar{X} - \mu))^T \right]$$

$$\sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T = \sum_{i=1}^{n} (X_i - \mu)(X_i - \mu)^T - \sum_{i=1}^{n} (X_i - \mu)(\bar{X} - \mu)^T - \sum_{i=1}^{n} (\bar{X} - \mu)(X_i - \mu)^T + \sum_{i=1}^{n} (\bar{X} - \mu)(\bar{X} - \mu)^T$$

$\sum_{i=1}^{n} (X_i - \mu) = n(\bar{X} - \mu)$

$$\sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T = \sum_{i=1}^{n} (X_i - \mu)(X_i - \mu)^T - n(\bar{X} - \mu)(\bar{X} - \mu)^T - n(\bar{X} - \mu)(\bar{X} - \mu)^T + n(\bar{X} - \mu)(\bar{X} - \mu)^T$$


$$\sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T = \sum_{i=1}^{n} (X_i - \mu)(X_i - \mu)^T - n(\bar{X} - \mu)(\bar{X} - \mu)^T$$

$$\mathbb{E}\left[ \sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T \right] = \sum_{i=1}^{n} \mathbb{E}\left[ (X_i - \mu)(X_i - \mu)^T \right] - n \mathbb{E}\left[ (\bar{X} - \mu)(\bar{X} - \mu)^T \right]$$

By definition:

$\mathbb{E}\left[ (X_i - \mu)(X_i - \mu)^T \right] = \Sigma$ (The population covariance matrix).

For the sample mean, $\text{Var}(\bar{X}) = \mathbb{E}\left[ (\bar{X} - \mu)(\bar{X} - \mu)^T \right] = \frac{1}{n}\Sigma$.

$$\mathbb{E}\left[ \sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T \right] = \sum_{i=1}^{n} \Sigma - n \left( \frac{1}{n}\Sigma \right)$$

$$\mathbb{E}\left[ \sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T \right] = n\Sigma - \Sigma = (n - 1)\Sigma$$

$\hat{\Sigma}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T$

$$\mathbb{E}[\hat{\Sigma}_{\text{MLE}}] = \frac{1}{n} \mathbb{E}\left[ \sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T \right] = \frac{n - 1}{n}\Sigma$$

This proves that $\hat{\Sigma}_{\text{MLE}}$ is inherently biased downward in finite samples, as $\frac{n-1}{n} < 1$.

Application of Bessel's Correction

To find an unbiased estimator $S$, we need $\mathbb{E}[S] = \Sigma$.

Given that $\mathbb{E}[\hat{\Sigma}_{\text{MLE}}] = \frac{n-1}{n}\Sigma$, we can isolate $\Sigma$ by multiplying both sides by the reciprocal scaling factor $\frac{n}{n-1}$:$$\Sigma = \frac{n}{n - 1} \mathbb{E}[\hat{\Sigma}_{\text{MLE}}]$$

Because expectation is a linear operator, we can move the scalar inside:$$\Sigma = \mathbb{E}\left[ \frac{n}{n - 1} \hat{\Sigma}_{\text{MLE}} \right]$$

Thus, our unbiased estimator $S$ is uniquely defined as:

$$S = \frac{n}{n - 1} \hat{\Sigma}_{\text{MLE}} = \frac{n}{n - 1} \left( \frac{1}{n} \sum_{i=1}^{n} (X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T \right) = \frac{1}{n - 1} \sum_{i=1}^{n} (X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T$$

This mathematically demonstrates why the denominator changes from $n$ to $n-1$ to preserve unbiasedness.

Definition: The probability of failing to reject $H_0$ when $H_1$ is true.

SHM Context: A missed detection. The diagnostic system treats a structurally compromised asset as healthy. This allows progressive damage to propagate unmonitored, potentially causing catastrophic structural failure.

Engineering Trade-offs and the Confidence Ellipsoid

Impact on Statistical Power

Statistical power is defined mathematically as $1 - \beta$ (the probability of correctly detecting damage when it exists).When an engineer reduces the false alarm rate to an ultra-conservative $\alpha = 0.0001$, the threshold for rejecting the null hypothesis is pushed far into the tails of the $H_0$ distribution. Because $\alpha$ and $\beta$ are inversely linked for a fixed sample size $n$, decreasing $\alpha$ dramatically increases $\beta$. Consequently, the statistical power ($1 - \beta$) plunges. The diagnostic layer becomes highly insensitive to early-stage structural degradation.

In multivariate feature space $\mathbb{R}^m$, the boundary of "healthy operation" is mathematically defined by a Mahalanobis distance threshold, corresponding to a confidence ellipsoid centered at $\hat{\mu}_n$:$$(X - \hat{\mu}_n)^T S^{-1} (X - \hat{\mu}_n) \leq \chi^2_{m} (1 - \alpha)$$Where $\chi^2_{m} (1 - \alpha)$ is the critical value of the Chi-squared distribution with $m$ degrees of freedom at a confidence level of $1 - \alpha$.

Geometric Modification: Setting $\alpha = 0.0001$ shifts the confidence level to $99.99\%$. This causes the critical threshold $\chi^2_{m} (0.9999)$ to scale outward sharply.

The Result: The geometric volume of the "healthy operation" confidence ellipsoid expands significantly. The acceptance region swallows up borderline anomalies and early damage signals, classifying them as nominal noise rather than a structural shift.












## **Part B: Theoretical Extension (Asymptotic Distributions & Slutsky's Theorem)**

The Lindeberg-Lévy Multivariate Central Limit Theorem establishes that as the sample snapshot size approaches infinity ($n \to \infty$), the standardized difference between the empirical mean and population mean converges in distribution to a Multivariate Gaussian:

$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$$



However, the true population covariance matrix $\boldsymbol{\Sigma}$ is a latent, unobservable parameter.

$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left({\boldsymbol{\mu}}, \frac{1}{n}\mathbf{S}\right)$$







1. Rigorous Statement of Slutsky's Theorem

Let $X_n$ and $Y_n$ be sequences of random vectors (or matrices) elements of $\mathbb{R}^d$ defined on a common probability space.

If $X_n$ converges in distribution to a random variable $X$ (denoted $X_n \xrightarrow{d} X$), and $Y_n$ converges in probability to a constant matrix/element $c$ (denoted $Y_n \xrightarrow{p} c$), then:

Addition: $X_n + Y_n \xrightarrow{d} X + c$

Multiplication: $Y_n X_n \xrightarrow{d} cX$ (provided the dimensions match for matrix multiplication)

Inverse/Division: $Y_n^{-1} X_n \xrightarrow{d} c^{-1}X$ (provided $c$ is invertible and $Y_n$ is invertible with probability approaching 1)

2. Proof of Substitution using Slutsky's Theorem

Establish the Component Limits

The Multivariate Central Limit Theorem (MCLT) states that:$$\sqrt{n}(\hat{\mu}_n - \mu) \xrightarrow{d} \mathcal{N}(0, \Sigma)$$

Let's standardize this vector expression by multiplying by the inverse of the true population Cholesky factor or square root matrix, $\Sigma^{-1/2}$:$$\Sigma^{-1/2} \left[ \sqrt{n}(\hat{\mu}_n - \mu) \right] \xrightarrow{d} \mathcal{N}(0, I_m)$$

Where $I_m$ is the $m \times m$ identity matrix. Let this sequence of random vectors be defined as $X_n = \sqrt{n}\Sigma^{-1/2}(\hat{\mu}_n - \mu)$, which converges to $X \sim \mathcal{N}(0, I_m)$.

Asymptotic Convergence of $S$

By the Weak Law of Large Numbers (WLLN), the empirical sample covariance matrix $S$ is a consistent estimator for the true population parameter $\Sigma$. Therefore, as $n \to \infty$:$$S \xrightarrow{p} \Sigma$$Because the matrix square root and matrix inversion operators are continuous functions on the space of positive-definite matrices, the Continuous Mapping Theorem (CMT) preserves this convergence:$$S^{-1/2} \xrightarrow{p} \Sigma^{-1/2}$$Let this sequence of matrices be defined as our Slutsky multiplier, $Y_n = S^{-1/2}$. It converges in probability to the constant matrix $c = \Sigma^{-1/2}$.

Application of Slutsky’s Theorem

Now, we construct the operational test statistic where the unknown $\Sigma$ is replaced by the sample-derived $S$:$$\mathcal{Z}_n = S^{-1/2} \left[ \sqrt{n}(\hat{\mu}_n - \mu) \right]$$We rewrite this expression to separate our $X_n$ and $Y_n$ components:$$\mathcal{Z}_n = \left( S^{-1/2} \Sigma^{1/2} \right) \left[ \sqrt{n}\Sigma^{-1/2}(\hat{\mu}_n - \mu) \right]$$

Applying Slutsky's Multiplication Rule ($Y_n X_n \xrightarrow{d} cX$):

The scalar/matrix multiplier sequence converges in probability: $\left( S^{-1/2} \Sigma^{1/2} \right) \xrightarrow{p} \left( \Sigma^{-1/2} \Sigma^{1/2} \right) = I_m$.

The random vector sequence converges in distribution: $\left[ \sqrt{n}\Sigma^{-1/2}(\hat{\mu}_n - \mu) \right] \xrightarrow{d} \mathcal{N}(0, I_m)$.Therefore, the combined product converges exactly to:$$\mathcal{Z}_n \xrightarrow{d} I_m \cdot \mathcal{N}(0, I_m) = \mathcal{N}(0, I_m)$$

We have proved that as $n \to \infty$:$$S^{-1/2} \left[ \sqrt{n}(\hat{\mu}_n - \mu) \right] \xrightarrow{d} \mathcal{N}(0, I_m)$$By shifting the scaling matrices back to the distribution side, we can express the asymptotic distribution of the empirical mean vector itself. Unrolling the standardization:$$\sqrt{n}(\hat{\mu}_n - \mu) \xrightarrow{d} \mathcal{N}(0, S)$$Dividing across the distributional boundary by $\sqrt{n}$ (which scales the covariance matrix by $1/n$):$$\hat{\mu}_n - \mu \xrightarrow{d} \mathcal{N}\left(0, \frac{1}{n}S\right)$$Which yields the final, practically applicable parametric modeling distribution used for real-time engineering diagnostics:$$\hat{\mu}_n \sim \mathcal{N}\left(\mu, \frac{1}{n}S\right)$$This concludes the formal proof. Slutsky's theorem guarantees that using $S$ instead of $\Sigma$ introduces error that vanishes entirely in the limit, ensuring your confidence ellipsoids remain asymptotically valid.

````python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

# Generating base stationary multivariate normal noise
base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    X = df.to_numpy()
    n, m = X.shape
    n_j = n // g_chunks

    # 1. Split df into g equal consecutive blocks
    chunks = [X[i * n_j : (i + 1) * n_j] for i in range(g_chunks)]

    # 2. Compute the global mean and individual chunk means
    global_mean = np.mean(X, axis=0)
    chunk_means = [np.mean(chunk, axis=0) for chunk in chunks]

    # 3. Calculate Within-Chunk Matrix (W) and Between-Chunk Matrix (B)
    W = np.zeros((m, m))
    B = np.zeros((m, m))

    for j in range(g_chunks):
        # Within-chunk variation
        diff_W = chunks[j] - chunk_means[j]
        W += diff_W.T @ diff_W

        # Between-chunk variation
        diff_B = chunk_means[j] - global_mean
        B += n_j * np.outer(diff_B, diff_B)

    # 4. Compute Wilks' Lambda, Bartlett's Chi-Square, and the final p-value
    wilks_lambda = np.linalg.det(W) / np.linalg.det(B + W)

    # Bartlett's transformation multiplier: -(n - 1 - (m + g) / 2)
    multiplier = -(n - 1 - (m + g_chunks) / 2)
    chi2_calc = multiplier * np.log(wilks_lambda)

    # Degrees of freedom: m * (g - 1)
    dof = m * (g_chunks - 1)
    p_value = 1.0 - stats.chi2.cdf(chi2_calc, dof)

    return {
        "Wilks' Lambda (Λ)": float(wilks_lambda),
        "Bartlett Chi-Square": float(chi2_calc),
        "Degrees of Freedom": int(dof),
        "p-value": float(p_value)
    }

# Execute completed function
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)

for k, v in results.items():
    print(f"{k}: {v:.6f}" if isinstance(v, float) else f"{k}: {v}")
````

1. Proof of Covariance Diagonalization

We are given the centered vector $\tilde{X}_i = (X_i - \mu) \in \mathbb{R}^m$, the population covariance matrix $\Sigma = E[\tilde{X}_i \tilde{X}_i^T]$, its spectral decomposition $\Sigma = P \Lambda P^T$, and the orthogonal transformation $Z_i = P^T \tilde{X}_i$.To find the theoretical covariance matrix of the transformed vector, $E[Z_i Z_i^T]$, we expand the expectation using linear operators:$$E[Z_i Z_i^T] = E\left[(P^T \tilde{X}_i)(P^T \tilde{X}_i)^T\right]$$Using the matrix transpose property $(AB)^T = B^T A^T$, we rewrite the internal term:$$E[Z_i Z_i^T] = E\left[P^T \tilde{X}_i \tilde{X}_i^T P\right]$$Because $P$ is a deterministic matrix of eigenvectors derived from the population distribution, it behaves as a constant with respect to the expectation operator $E[\cdot]$. Thus, we can factor $P^T$ out to the left and $P$ out to the right:$$E[Z_i Z_i^T] = P^T E\left[\tilde{X}_i \tilde{X}_i^T\right] P$$Substituting the definition of the baseline covariance matrix $\Sigma = E[\tilde{X}_i \tilde{X}_i^T]$:$$E[Z_i Z_i^T] = P^T \Sigma P$$Now, substitute the spectral decomposition $\Sigma = P \Lambda P^T$:$$E[Z_i Z_i^T] = P^T (P \Lambda P^T) P = (P^T P) \Lambda (P^T P)$$Because $\Sigma$ is symmetric, its eigenvectors form an orthonormal basis, meaning $P$ is an orthogonal matrix satisfying $P^T P = P P^T = I_m$. Substituting the identity matrix $I_m$:$$E[Z_i Z_i^T] = I_m \Lambda I_m = \Lambda$$Since $\Lambda = \text{diag}(\lambda_1, \lambda_2, \dots, \lambda_m)$, the matrix $E[Z_i Z_i^T]$ simplifies exactly to a strictly diagonal matrix. $\blacksquare$

2. Geometric and Statistical Meaning of Diagonalization

Statistical Meaning

The fact that $E[Z_i Z_i^T] = \Lambda$ is strictly diagonal implies that the cross-covariance between any two distinct coordinates $Z_{i,j}$ and $Z_{i,k}$ (where $j \neq k$) is exactly zero:$$\text{Cov}(Z_{i,j}, Z_{i,k}) = 0 \quad \forall j \neq k$$Statistically, the transformation completely decorrelates the features. In the transformed principal component space, knowing the variation of a sample along the $j$-th coordinate provides zero linear information about its position along the $k$-th coordinate.

Geometric Meaning

Geometrically, the orthogonal matrix $P^T$ acts as a rigid rotation (and potentially a reflection) of the original coordinate axes. The original data cloud may have been highly skewed, forming an oblique hyper-ellipsoid where the axes of maximum variation were coupled across multiple sensor dimensions.The mapping aligns the new coordinate axes ($Z$) exactly with the principal axes (major and minor axes) of this hyper-ellipsoid. The diagonal elements $\lambda_j$ represent the physical variances (spread) along these newly aligned, mutually perpendicular spatial directions.

3. Proof of Total Variance Preservation

The total variance of a multivariate system is defined as the sum of the variances of its individual scalar channels, which corresponds to the trace ($\text{tr}$) of its covariance matrix.We begin with the trace of the transformed covariance matrix $\Lambda$:$$\text{tr}(\Lambda) = \text{tr}(P^T \Sigma P)$$The trace operator enjoys a cyclic permutation property: for any matrices $A$ and $B$ such that the products are square, $\text{tr}(AB) = \text{tr}(BA)$. Let $A = P^T \Sigma$ and $B = P$. Applying the cyclic shift:$$\text{tr}((P^T \Sigma) P) = \text{tr}(P (P^T \Sigma))$$Using associative grouping:$$\text{tr}(P P^T \Sigma) = \text{tr}((P P^T) \Sigma)$$Because $P$ is orthogonal, $P P^T = I_m$. Therefore:$$\text{tr}(I_m \Sigma) = \text{tr}(\Sigma)$$Thus, $\text{tr}(\Sigma) = \text{tr}(\Lambda)$. Because $\Lambda$ is a diagonal matrix, its trace is explicitly the sum of its diagonal elements:$$\text{tr}(\Sigma) = \text{tr}(\Lambda) = \sum_{j=1}^{m} \lambda_j$$This proves that the total kinetic/statistical energy of the system is perfectly conserved when moving from the raw physical sensor space to the decoupled principal space. $\blacksquare$

4. Subspace Variance Formulations

When the empirical matrix is truncated to the first $k$ principal components, we partition our eigenvalues into an explained set ($\lambda_1 \dots \lambda_k$) and an unexplained set ($\lambda_{k+1} \dots \lambda_m$).

Cumulative Explained Variance Ratio

$\Phi(k)$The proportion of total system variation captured within the dominant $k$-dimensional hyperplane is given by:$$\Phi(k) = \frac{\text{tr}(\hat{P}_k^T \Sigma \hat{P}_k)}{\text{tr}(\Sigma)} = \frac{\sum_{j=1}^{k} \lambda_j}{\sum_{j=1}^{m} \lambda_j}$$

Residual Unexplained Variance Ratio

$\Psi(k)$The proportion of variation cast away into the orthogonal residual subspace is given by:$$\Psi(k) = \frac{\text{tr}(\hat{P}_{m-k}^T \Sigma \hat{P}_{m-k})}{\text{tr}(\Sigma)} = \frac{\sum_{j=k+1}^{m} \lambda_j}{\sum_{j=1}^{m} \lambda_j} = 1 - \Phi(k)$$

5. Subspace Energy Decomposition (Pythagorean Identity)

$$(x_i - \hat{\mu}_n) = \hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k}$$Where the reconstructed vector is $\hat{x}_i = \hat{\mu}_n + \hat{P}_k z_{i,k}$ and the residual error vector is $e_i = \hat{P}_{m-k} z_{i,m-k}$.

$$(x_i - \hat{\mu}_n) = \hat{x}_i - \hat{\mu}_n + e_i$$

$$\|x_i - \hat{\mu}_n\|^2 = (x_i - \hat{\mu}_n)^T (x_i - \hat{\mu}_n)$$

$$\|x_i - \hat{\mu}_n\|^2 = (\hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k})^T (\hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k})$$

$$\|x_i - \hat{\mu}_n\|^2 = (z_{i,k}^T \hat{P}_k^T + z_{i,m-k}^T \hat{P}_{m-k}^T) (\hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k})$$

$$\|x_i - \hat{\mu}_n\|^2 = z_{i,k}^T \hat{P}_k^T \hat{P}_k z_{i,k} + z_{i,k}^T \hat{P}_k^T \hat{P}_{m-k} z_{i,m-k} + z_{i,m-k}^T \hat{P}_{m-k}^T \hat{P}_k z_{i,k} + z_{i,m-k}^T \hat{P}_{m-k}^T \hat{P}_{m-k} z_{i,m-k}$$

By the properties of spectral partitioning, the full eigenvector matrix $P = [\hat{P}_k \mid \hat{P}_{m-k}]$ is orthonormal. This forces two strict geometric conditions:

Subspace Orthonormal Bases: $\hat{P}_k^T \hat{P}_k = I_k$ and $\hat{P}_{m-k}^T \hat{P}_{m-k} = I_{m-k}$

Mutual Subspace Orthogonality: $\hat{P}_k^T \hat{P}_{m-k} = \mathbf{0}_{k \times (m-k)}$ and $\hat{P}_{m-k}^T \hat{P}_k = \mathbf{0}_{(m-k) \times k}$

$$\|x_i - \hat{\mu}_n\|^2 = z_{i,k}^T (I_k) z_{i,k} + 0 + 0 + z_{i,m-k}^T (I_{m-k}) z_{i,m-k}$$

$$\|x_i - \hat{\mu}_n\|^2 = z_{i,k}^T z_{i,k} + z_{i,m-k}^T z_{i,m-k}$$

Since $v^T v = \|v\|^2$ for any real vector, this leaves us precisely with:$$\|x_i - \hat{\mu}_n\|^2 = \|z_{i,k}\|^2 + \|z_{i,m-k}\|^2$$This proves that total raw observation energy decomposes cleanly and independently into explained principal energy and unexplained residual energy. $\blacksquare$

6. Diagnostic Contrasts in Structural Health Monitoring (SHM)

Comparison of Diagnostic Functions

Hotelling’s $T^2$ Statistic: Measures variations within the normal operational subspace (the hyperplane spanned by $\hat{P}_k$). It tracks Mahalanobis distance from the baseline center, monitoring whether the normal operational characteristics have scaled up or shifted in amplitude along expected directions.

Residual $Q$ Statistic (Squared Prediction Error - SPE): Measures variations outside the normal operational subspace (the residual hyperplane spanned by $\hat{P}_{m-k}$). It tracks the perpendicular distance of an observation away from the PCA hyperplane, monitoring whether the underlying correlation structure between the sensors has fundamentally changed.



## **Part B: Numerical Verification**



***Scenario:** You are building a real-time diagnostics layer for a structural component monitored by $m = 4$ sensors. You have a baseline archive of $n = 3,000$ nominal snapshots. To establish operational boundaries, you must write an optimization pipeline that analyzes how Hotelling’s $T^2$ and the residual $Q$ statistic respond to different truncation boundaries $k \in \{1, 2, 3\}$.*

**Your Task:**
Complete the Python class method below inside Google Colab. The function must:

1. Standardize and center the incoming realizations.
2. Decompose the unbiased sample covariance matrix $\mathbf{S}$ using `np.linalg.eigh` and sort components in descending order ($\widehat{\lambda}_1 \ge \dots \ge \widehat{\lambda}_m$).
3. Loop through every possible subspace cutoff value $k \in \{1, 2, 3\}$, calculating the sample vector array for Hotelling’s $T^2$ ($T^2_i = \sum_{j=1}^k \frac{z_{i,j}^2}{\widehat{\lambda}_j}$) and the $Q$ statistic ($Q_i = \sum_{j=k+1}^m z_{i,j}^2$).
4. Compute the expected value (empirical mean) of $T^2$ and $Q$ across all samples for each $k$.
5. Return these metrics to verify your theoretical derivations in Part A.



````python
import numpy as np
import pandas as pd

# --- Synthetic Structural Snapshot Generator ---
np.random.seed(101)
n_snapshots = 3000
m_sensors = 4

# Simulating structural baseline data with coupled operational variations
latent_factor = np.random.normal(0, 1.5, size=(n_snapshots, 1))
sensor_noise = np.random.multivariate_normal(
    mean=[0, 0, 0, 0],
    cov=np.diag([0.1, 0.15, 0.08, 0.12]),
    size=n_snapshots
)
# Mixing matrix mapping a dominant operational load across channels
mixing_weights = np.array([[0.8], [1.2], [-0.5], [1.5]])
raw_snapshots = latent_factor @ mixing_weights.T + sensor_noise

df_shm = pd.DataFrame(raw_snapshots, columns=['Sensor_A', 'Sensor_B', 'Sensor_C', 'Sensor_D'])
# ----------------------------------------------------------------

class SHMDiagnosticPipeline:
    def __init__(self, data: pd.DataFrame):
        self.df = data
        self.n, self.m = data.shape
        self.X_scaled = None
        self.eigenvalues = None
        self.eigenvectors = None
        self.Z = None

    def fit_baseline_space(self):
        """
        Centers and standardizes the dataset, computes the unbiased sample
        covariance matrix, and executes sorted spectral decomposition.
        """
        # 1. Standardize and center the incoming realizations (Z-score)
        mean = self.df.mean(axis=0).to_numpy()
        std = self.df.std(axis=0).to_numpy()  # ddof=1 by default
        self.X_scaled = (self.df.to_numpy() - mean) / std

        # 2. Decompose the unbiased sample covariance matrix S (ddof=1)
        S = (self.X_scaled.T @ self.X_scaled) / (self.n - 1)

        # np.linalg.eigh returns eigenvalues in ascending order
        evals, evecs = np.linalg.eigh(S)

        # Sort components in strictly descending order
        idx = np.argsort(evals)[::-1]
        self.eigenvalues = evals[idx]
        self.eigenvectors = evecs[:, idx]

        # Project centered realizations into principal component space coordinates
        self.Z = self.X_scaled @ self.eigenvectors

    def evaluate_truncation_boundaries(self) -> dict:
        """
        Loops through every possible subspace cutoff value k, calculating
        Hotelling's T^2 and residual Q arrays, and extracting empirical means.
        """
        if self.Z is None:
            self.fit_baseline_space()

        summary_metrics = {}

        # Loop through every possible subspace cutoff value k from 1 to m-1
        for k in range(1, self.m):
            # T^2 component: Sum of squared principal coordinates weighted by inverse eigenvalues
            # z_i,j^2 / lambda_j across columns 0 to k-1
            t2_samples = np.sum((self.Z[:, :k] ** 2) / self.eigenvalues[:k], axis=1)

            # Q component: Sum of squared principal coordinates in the residual subspace
            # z_i,j^2 across columns k to m-1
            q_samples = np.sum(self.Z[:, k:] ** 2, axis=1)

            # Compute expected value (empirical mean) across all samples
            mean_t2 = np.mean(t2_samples)
            mean_q = np.mean(q_samples)

            summary_metrics[f"k={k}"] = {
                "Expected_T2": float(mean_t2),
                "Expected_Q": float(mean_q)
            }

        return summary_metrics

# Execute the diagnostic evaluation pipeline
pipeline = SHMDiagnosticPipeline(df_shm)
diagnostic_results = pipeline.evaluate_truncation_boundaries()

# Print formatted verification summary
print("--- SHM Subspace Truncation Metric Summary ---")
for cutoff, metrics in diagnostic_results.items():
    print(f"\nSubspace Boundary Layer [{cutoff}]:")
    print(f"  -> E[T²] (Empirical Mean): {metrics['Expected_T2']:.6f}")
    print(f"  -> E[Q]  (Empirical Mean): {metrics['Expected_Q']:.6f}")
````

Hotelling's $T^2$ Expectation: Because the data is standardized, each squared principal component $z_{i,j}^2$ divided by its corresponding eigenvalue $\lambda_j$ creates a collection of independent, standard unit-variance variables. By definition, the expected value of each individual term in the sum is $E[z_{i,j}^2 / \lambda_j] = 1$. Therefore, for a subspace cutoff $k$, the expected value scales exactly to the degrees of freedom utilized:$$E[T^2] = k$$

Residual $Q$ Expectation: The residual $Q$ statistic aggregates the remaining unscaled variance left over in the discarded dimensions. Because the matrix trace (total variance) of our standardized system is perfectly preserved and equals the total number of dimensions ($m=4$), discarding the columns from $k$ to $m$ leaves exactly $m-k$ active channels. Since the average value of unscaled variance matches the sum of the remaining eigenvalues, its empirical mean simplifies cleanly to:$$E[Q] = m - k$$

Thus, at every boundary, $E[T^2] + E[Q] = m = 4.000000$, validating that total space energy is rigidly conserved during truncation.

Part 1: Proof of the Fundamental Equation of Factor Analysis

To derive the Fundamental Equation of Factor Analysis, we begin with the linear generative structure for a normalized, zero-mean baseline feature vector $Z_i \in \mathbb{R}^m$:$$Z_i = \Lambda F_i + \epsilon_i$$The population covariance matrix $R$ is defined as the expected value of the outer product of $Z_i$ with itself:$$R = \mathbb{E}[Z_i Z_i^T]$$Substituting the generative structure into this expectation yields:$$R = \mathbb{E}\left[(\Lambda F_i + \epsilon_i)(\Lambda F_i + \epsilon_i)^T\right]$$Expanding the terms inside the expectation:$$R = \mathbb{E}\left[(\Lambda F_i + \epsilon_i)(F_i^T \Lambda^T + \epsilon_i^T)\right]$$$$R = \mathbb{E}\left[\Lambda F_i F_i^T \Lambda^T + \Lambda F_i \epsilon_i^T + \epsilon_i F_i^T \Lambda^T + \epsilon_i \epsilon_i^T\right]$$Using the linearity of the expectation operator, we can distribute $\mathbb{E}[\cdot]$ across the individual terms:$$R = \Lambda \mathbb{E}[F_i F_i^T] \Lambda^T + \Lambda \mathbb{E}[F_i \epsilon_i^T] + \mathbb{E}[\epsilon_i F_i^T] \Lambda^T + \mathbb{E}[\epsilon_i \epsilon_i^T]$$

Orthogonal common factors: $\mathbb{E}[F_i F_i^T] = I_{k \times k}$

Independent latent factors and errors: $\mathbb{E}[\epsilon_i F_i^T] = 0 \implies \mathbb{E}[F_i \epsilon_i^T] = 0$

Diagonalized measurement noise matrix: $\mathbb{E}[\epsilon_i \epsilon_i^T] = \Psi = \text{diag}(\varphi_1^2, \dots, \varphi_m^2)$

$$R = \Lambda I_{k \times k} \Lambda^T + \Lambda(0) + (0)\Lambda^T + \Psi$$$$R = \Lambda \Lambda^T + \Psi$$

This yields the explicit definitions:

Uniqueness ($\varphi_j^2$)$$\varphi_j^2 = 1 - h_j^2$$

In an anomaly detection matrix, a sudden spike in a sensor's residual variance that deviates from $\varphi_j^2$ signals a point-failure (e.g., a broken sensor wire or localized probe fault). Conversely, a shift tracking along the communality subspace $h_j^2$ indicates a shared physical process anomaly, such as a bearing failure manifesting across multiple thermal and vibration channels simultaneously.

Part 3: Orthogonal Coordinate Rotations (Varimax)

Raw PCA Loading Vector: Driven purely by geometric variance maximization. The first loading vector aligns with the direction of maximal global variance in the data space. Consequently, raw PCA components are often general factors where nearly all sensors exhibit moderate-to-high loadings, making physical isolation difficult.

Varimax-Rotated Factor Loading Vector: Driven by a structural search for a "simple structure." The Varimax criterion mathematically maximizes the variance of the squared loadings within the columns of $\Lambda$. Visually, it rotates the factor axes so that each sensor loads heavily onto only one (or a very few) latent factors, and near zero on others. This cleanly maps specific groups of sensors to distinct physical phenomena.

Invariance of Communality and Global Covariance

Let $T$ be an orthogonal rotation matrix such that $T T^T = T^T T = I_{k \times k}$. The rotated loading matrix is defined as $\tilde{\Lambda} = \Lambda T$.We substitute the rotated loadings back into the global covariance approximation:$$\tilde{\Lambda} \tilde{\Lambda}^T + \Psi = (\Lambda T)(\Lambda T)^T + \Psi = \Lambda (T T^T) \Lambda^T + \Psi$$Because $T T^T = I$, this simplifies exactly to:$$\Lambda I \Lambda^T + \Psi = \Lambda \Lambda^T + \Psi$$Thus, the global covariance approximation $R$ remains completely unchanged.For the individual sensor communality, the $j$-th diagonal entry of the matrix product $\tilde{\Lambda}\tilde{\Lambda}^T$ is equivalent to the row-norm of the $j$-th row of $\tilde{\Lambda}$. Let $\lambda_{j\cdot}$ be the $j$-th row vector of $\Lambda$. The rotated row vector is $\tilde{\lambda}_{j\cdot} = \lambda_{j\cdot} T$. The communality is the inner product of this row with itself:$$\tilde{h}_j^2 = \tilde{\lambda}_{j\cdot} \tilde{\lambda}_{j\cdot}^T = (\lambda_{j\cdot} T)(\lambda_{j\cdot} T)^T = \lambda_{j\cdot} (T T^T) \lambda_{j\cdot}^T = \lambda_{j\cdot} \lambda_{j\cdot}^T = h_j^2$$Because the length of a row vector is invariant under orthogonal transformations, individual sensor communalities are preserved exactly.

## **Part B: Parametric Factor Partition Engine & UI Architecture**

The objective is to implement an end-to-end data processing and diagnostic visualization pipeline that instantiates a parametric **Factor Analysis (FA)** engine with **Varimax rotation**. The engine must partition empirical observation metrics into a shared structural subspace and a distinct diagonal noise floor, mapping the results onto a highly structured, asymmetric grid interface.

### Develop a Mathematical Engine & Pre-processing Guardrails


Implement an isolated algorithm to structure and decompose empirical data structures according to the following guidelines:

1. **Z-Score Mapping Layer:** Accept an arbitrary data frame and isolate numeric observation columns. Standardize the observations into standard normal space $Z \in \mathbb{R}^{n \times m}$ applying an unbiased sample standard deviation ($ddof=1$). Build an explicit numerical guardrail that catches zero-variance channels ($\sigma_j = 0$) and caps them to a minimum boundary threshold of $10^{-15}$ to prevent divide-by-zero runtime exceptions.
2. **Dimensionality Constraint Layer:** Let $m$ represent the count of isolated sensor features, and $k$ represent the target count of latent physical factors. Program two validation rule exceptions:
* Reject processing if the operational snapshot count ($n$) is less than or equal to the channel count ($m$).
* Throw an descriptive error if the requested latent space dimension ($k$) is not strictly less than the available feature space dimension ($m$).


3. **Information Decomposition Layer:** Fit a Factor Analysis framework utilizing a Maximum Likelihood or Singular Value Decomposition (SVD) algorithm combined with an orthogonal Varimax rotation matrix ($T$). From the fitted parameters, compute the vector arrays for:
* **Communality Vector ($h^2 \in \mathbb{R}^m$):** The total shared variance captured across each channel.
* **Uniqueness Vector ($\varphi^2 \in \mathbb{R}^m$):** The remaining isolated sensor noise floor.

````python
import numpy as np
import pandas as pd


class MathematicalEngine:
    """A mathematical engine for data standardization, validation guardrails,

    and latent factor decomposition.
    """

    def __init__(self, target_factors: int, epsilon: float = 1e-15):
        """Initialize the engine with configuration parameters.

        :param target_factors: (k) Target count of latent physical factors.
        :param epsilon: Minimum boundary threshold for zero-variance channels.
        """
        self.k = target_factors
        self.epsilon = epsilon

    def z_score_mapping_layer(self, df: pd.DataFrame) -> np.ndarray:
        """Isolates numeric columns, standardizes observations using unbiased sample

        standard deviation (ddof=1), and guards against zero-variance channels.
        """
        # Isolate numeric observation columns
        numeric_df = df.select_dtypes(include=[np.number])
        if numeric_df.empty:
            raise ValueError(
                "Dataframe contains no numeric observation columns."
            )

        X = numeric_df.to_numpy(dtype=np.float64)

        # Compute empirical mean and unbiased sample standard deviation
        mean = np.mean(X, axis=0)
        std = np.std(X, axis=0, ddof=1)

        # Numerical Guardrail: Catch zero-variance channels (std == 0) and cap them
        std_safe = np.where(std == 0, self.epsilon, std)

        # Standardize to Standard Normal Space Z
        Z = (X - mean) / std_safe
        return Z

    def enforce_dimensionality_constraints(self, Z: np.ndarray) -> None:
        """Validates structural shape constraints on the operational snapshot matrix.

        :param Z: Standardized data matrix of shape (n, m)
        """
        n, m = Z.shape

        # Rule 1: Reject processing if snapshot count (n) <= channel count (m)
        if n <= m:
            raise ValueError(
                f"Dimensionality Breach: Snapshot count n={n} must be greater than "
                f"channel count m={m} to prevent rank deficiency."
            )

        # Rule 2: Throw descriptive error if latent space (k) is not strictly less than feature space (m)
        if self.k >= m:
            raise ValueError(
                f"Constraint Violation: Requested latent space dimension k={self.k} "
                f"must be strictly less than available feature space dimension m={m}."
            )

    @staticmethod
    def _varimax_rotation(loadings: np.ndarray, max_iter: int = 500, tol: float = 1e-6) -> np.ndarray:
        """Applies an orthogonal Varimax rotation matrix (T) to the factor loadings."""
        X = loadings
        p, k = X.shape
        if k < 2:
            return X

        # Initialize rotation matrix T as identity
        T = np.eye(k)
        d = 0

        for _ in range(max_iter):
            d_old = d
            # Compute current rotated loadings
            Lambda = np.dot(X, T)
            # Varimax objective function gradient component
            u, s, vh = np.linalg.svd(
                np.dot(X.T, Lambda**3 - (1.0 / p) * np.dot(Lambda, np.diag(np.sum(Lambda**2, axis=0))))
            )
            T = np.dot(u, vh)
            d = np.sum(s)
            if d_old != 0 and (d - d_old) / d_old < tol:
                break

        return np.dot(X, T)

    def information_decomposition_layer(self, Z: np.ndarray) -> dict:
        """Fits a Factor Analysis architecture using Singular Value Decomposition (SVD)

        and extracts structural variance components.
        """
        n, m = Z.shape

        # Compute SVD of the scaled data matrix (Z / sqrt(n - 1))
        # This matches the principal component extraction approach for FA loadings
        _, s, Vt = np.linalg.svd(Z / np.sqrt(n - 1), full_matrices=False)

        # Extract unrotated loadings for the top k latent factors
        # Loadings = Right singular vectors scaled by singular values
        loadings = Vt[: self.k, :].T * s[: self.k]

        # Apply orthogonal Varimax rotation matrix (T)
        rotated_loadings = self._varimax_rotation(loadings)

        # Compute Communality Vector (h^2): total shared variance across each channel
        communality = np.sum(rotated_loadings**2, axis=1)

        # Compute Uniqueness Vector (phi^2): remaining isolated sensor noise floor
        # Since total variance of standardized Z channels is 1.0: phi^2 = 1 - h^2
        uniqueness = 1.0 - communality

        # Ensure values don't breach physical bounds due to floating-point noise
        communality = np.clip(communality, 0.0, 1.0)
        uniqueness = np.clip(uniqueness, 0.0, 1.0)

        return {
            "loadings": rotated_loadings,
            "communality": communality,
            "uniqueness": uniqueness,
        }

    def fit_transform(self, df: pd.DataFrame) -> dict:
        """Executes the pipeline sequentially through all guardrails and layers."""
        # 1. Z-Score Mapping Layer
        Z = self.z_score_mapping_layer(df)

        # 2. Dimensionality Constraint Layer
        self.enforce_dimensionality_constraints(Z)

        # 3. Information Decomposition Layer
        metrics = self.information_decomposition_layer(Z)

        return metrics
````

### Latent Score Projection

1. **Mathematical State Reconstruction:** Using Thomson’s Minimum Mean Squared Error (MMSE) regression methodology, transform the standard normal snapshot array ($Z$) directly into hidden factor states $F \in \mathbb{R}^{n \times k}$ using the system relation:

$$F = Z R^{-1} \Lambda_{\text{rotated}}$$


2. **Performance Diagnostic Telemetry:** Calculate the empirical variance of the reconstructed factor scores enforcing an unbiased scale correction ($ddof=1$). Print a clean confirmation breakdown out to the logging console that presents the calculated system-wide metrics:

$$\text{Average System Communality \%} = \frac{1}{m}\sum_{j=1}^m h_j^2 \times 100$$


$$\text{Average System Uniqueness \%} = \frac{1}{m}\sum_{j=1}^m \varphi_j^2 \times 100$$

````python
import numpy as np

def latent_score_projection(Z, Lambda_rotated):
    """
    Performs Latent Score Projection using Thomson's MMSE methodology
    and prints performance diagnostic telemetry.

    Parameters:
    -----------
    Z : numpy.ndarray
        Standard normal snapshot array of shape (n_samples, m_features)
    Lambda_rotated : numpy.ndarray
        Rotated factor loading matrix of shape (m_features, k_factors)
    """
    n, m = Z.shape
    _, k = Lambda_rotated.shape

    print("=" * 60)
    print("INITIALIZING LATENT SCORE PROJECTION TELEMETRY")
    print("=" * 60)

    # 1. Calculate the empirical correlation matrix R of observed variables
    # Since Z is standard normal, R = (Z^T * Z) / (n - 1)
    R = np.corrcoef(Z.T)

    # 2. Mathematical State Reconstruction (Thomson's MMSE)
    # F = Z * R^-1 * Lambda_rotated
    R_inv = np.linalg.inv(R)
    F = Z @ R_inv @ Lambda_rotated

    # 3. Performance Diagnostic Telemetry
    # Calculate empirical variance of reconstructed factor scores (ddof=1)
    factor_variances = np.var(F, axis=0, ddof=1)

    # Calculate Communalities (h_j^2) and Uniquenesses (phi_j^2)
    # For orthogonal/rotated factors, h_j^2 is the sum of squared loadings per variable
    communalities = np.sum(Lambda_rotated**2, axis=1)
    uniquenesses = 1.0 - communalities

    # System-wide metrics calculations
    avg_communality_pct = np.mean(communalities) * 100
    avg_uniqueness_pct = np.mean(uniquenesses) * 100

    # --- LOGGING CONSOLE OUTPUT ---
    print("\n[SYSTEM DIAGNOSTIC: FACTOR RECONSTRUCTION]")
    for idx, var in enumerate(factor_variances):
        print(f" -> Empirical Variance (Factor {idx+1}): {var:.4f}")

    print("\n[SYSTEM-WIDE METRICS BREAKDOWN]")
    print(f" * Average System Communality % : {avg_communality_pct:.2f}%")
    print(f" * Average System Uniqueness %  : {avg_uniqueness_pct:.2f}%")
    print("=" * 60)

    return F

# --- Demonstration with Synthetic Data ---
if __name__ == "__main__":
    np.random.seed(42)

    # Dimensions: n=1000 snapshots, m=6 observed variables, k=2 hidden factors
    n_samples = 1000
    m_features = 6
    k_factors = 2

    # Generate standard normal snapshot array Z
    Z_raw = np.random.normal(0, 1, size=(n_samples, m_features))
    Z_standardized = (Z_raw - np.mean(Z_raw, axis=0)) / np.std(Z_raw, axis=0)

    # Mock a rotated factor loading matrix Lambda (satisfying communalities <= 1)
    Lambda_mock = np.array([
        [0.81, 0.15],
        [0.72, 0.22],
        [0.10, 0.85],
        [0.05, 0.78],
        [0.60, 0.40],
        [0.45, 0.50]
    ])

    # Execute projection and telemetry
    F_states = latent_score_projection(Z_standardized, Lambda_mock)
````
